**Problem 1: Smart Light Controller**

```python
class SmartLight:
    def set_light(self, name, action):
        self.light_name = name
        self.status = action

    def display_status(self):
        print(f"{self.light_name} is {self.status}")



name = input()
action = input()


light = SmartLight()
light.set_light(name, action)
light.display_status()
```

**Problem 2: Employee ID Card System**

```python
class Employee:
    def set_details(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_details(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


name = input()
emp_id = input()
department = input()


emp = Employee()
emp.set_details(name, emp_id, department)
emp.display_details()
```

**Problem 3: Mobile Contact Record**

```python
class Contact:
    def set_contact(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")



name = input()
phone = input()


c = Contact()
c.set_contact(name, phone)
c.display_contact()
```

**Problem 4: Product Price Tag Generator**

```python
class Product:
    def set_product(self, name, price):
        self.name = name
        self.price = price

    def display_product(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")


name = input()
price = int(input())


p = Product()
p.set_product(name, price)
p.display_product()
```

**Problem 5: Movie Rating Display System**

```python
class Movie:
    def set_movie(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")



name = input()
rating = float(input())


m = Movie()
m.set_movie(name, rating)
m.display_movie()
```
