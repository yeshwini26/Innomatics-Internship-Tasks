**Problem 1: Smart Parking Lot Management** **System**

```python
def manage_parking(capacity, logs):
    current = 0
    peak = 0

    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT":
            current -= 1

        if current > peak:
            peak = current

    print("Currently Parked Vehicles:", current)
    print("Peak Parking Usage:", peak)

    if current > capacity:
        print("Parking Status: Full Alert!")
    else:
        print("Parking Status: Available")


# Sample Input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

manage_parking(capacity, logs)
```

**Problem 2: Online Food Delivery Time Estimator**

```python
def estimate_delivery(distance, traffic, weather):
    time = distance * 5

    if traffic == "Low":
        time += 5
    elif traffic == "Medium":
        time += 10
    elif traffic == "High":
        time += 20


    if weather == "Rainy":
        time += 15
    elif weather == "Storm":
        time += 25

    print("Estimated Delivery Time:", time, "minutes")



estimate_delivery(8, "High", "Rainy")
```

**Problem 3: Movie Theatre Seat Occupancy Analyzer**

```python
def analyze_occupancy(total_seats, booked_seats):

    booked = len(booked_seats)

    occupancy = (booked / total_seats) * 100

    print("Occupancy:", occupancy, "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")


# Sample Input
total_seats = 200
booked_seats = [1] * 150

analyze_occupancy(total_seats, booked_seats)
```

**Problem 4: Cloud Server Load Classification System**

```python
def classify_server_load(cpu_readings):

    total = 0

    for cpu in cpu_readings:
        total += cpu

    avg = total / len(cpu_readings)

    print("Average CPU Load:", avg, "%")

    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
readings = [45, 60, 70, 85, 90]

classify_server_load(readings)
```

**Problem 5: Smart Classroom Resource Usage Monitor**

```python
def monitor_resources(resources):

    overused = []

    for resource in resources:
        if resources[resource] > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(resources)
```

**Problem 6: Online Event Registration Capacity Controller**

```python
def manage_registrations(capacity, registrations):

    if registrations <= capacity:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


# Sample Input
manage_registrations(100, 105)
```
