**Problem 1: Mobile Recharge Validation System**

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount >= 50 and amount in valid_plans:
        return True
    return False

while True:
    amount = int(input("Enter recharge amount: "))
    if validate_recharge(amount):
        print("Recharge Successful")
        break
    else:
        print("Invalid Plan or Amount. Try Again.")
```

**Problem 2: Inventory Reorder Alert System**

```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")

products = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18
}

check_inventory(products)
```

**Problem 3: Student Result Processing System**

```python
def process_result(marks):
    total = 0
    for mark in marks:
        total += mark
    average = total / len(marks)
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

marks = list(map(int, input("Enter marks separated by space: ").split()))
result = process_result(marks)
print("Result:", result)
```

**Problem 4: Cab Fare Estimator with Retry Option**

```python
def calculate_fare(distance, peak):
    fare = 50 + (12 * distance)
    if peak.lower() == "yes":
        fare += fare * 0.25
    return fare

while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")
    fare = calculate_fare(distance, peak)
    print("Total Fare:", fare)
    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

**Problem 5: Employee Attendance Eligibility Checker**

```python
def check_eligibility(attendance):
    present = 0
    for day in attendance:
        if day == "P":
            present += 1
    percentage = (present / len(attendance)) * 100
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance = input("Enter attendance (P/A separated by space): ").split()
result = check_eligibility(attendance)
print("Attendance Status:", result)
```

**Problem 6: Password Strength Checker**

```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak Password"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

password = input("Enter password: ")
print(check_password(password))
```
